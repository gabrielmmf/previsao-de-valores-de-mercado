```python
import numpy as np
def dummie_age(df):
    df['lessThan20Years'] = 0
    df['21To24Years'] = 0
    df['26To30Years'] = 0
    df['MoreThan30Years'] = 0

    # Define os valores das dummies com base na idade
    df.loc[df['playerAge'] <= 20, 'lessThan20Years'] = 1
    df.loc[(df['playerAge'] >= 21) & (df['playerAge'] <= 25), '21To24Years'] = 1
    df.loc[(df['playerAge'] >= 26) & (df['playerAge'] <= 30), '26To30Years'] = 1
    df.loc[df['playerAge'] > 30, 'MoreThan30Years'] = 1

    return ['lessThan20Years', '21To24Years','26To30Years','MoreThan30Years']

import pycountry_convert as pc

def get_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    # adding manual continents for some countries
    except KeyError:
        if country_name in ['Wales', 'England', 'Scotland', 'Bosnia-Herzegovina', 'Northern Ireland', 'Kosovo']:
            return 'Europe'
        elif country_name in ["Cote d'Ivoire" ,'The Gambia', 'Curacao', 'DR Congo' ]:
            return 'Africa'
        elif country_name in ['Korea, South', 'Korea, North']:
            return 'Asia'
        elif country_name in ['Neukaledonien']:
            return 'Oceania'
        else:
            return None

def dummie_nationality(df):
    df['continent'] = df['playerCountry'].apply(get_continent)
    df['european'] = 0
    df['asian'] = 0
    df['southAmerican'] = 0
    df['african'] = 0
    df['northAmerican'] = 0
    df['oceanian'] = 0

    df.loc[df['continent'] == 'Europe', 'european'] = 1
    df.loc[df['continent'] == 'Asia', 'asian'] = 1
    df.loc[df['continent'] == 'South America', 'southAmerican'] = 1
    df.loc[df['continent'] == 'North America', 'northAmerican'] = 1
    df.loc[df['continent'] == 'Oceania', 'oceanian'] = 1

    return ['european', 'asian','southAmerican','african','northAmerican','oceanian']

season_average = {}
for season in detailed_transfers['season'].unique():
    season_average[season] = detailed_transfers.loc[detailed_transfers['season'] == season,'transferFee'].mean()
print(season_average)

transfer_fee = detailed_transfers['transferFee']
upper_threshold = np.percentile(transfer_fee, 95)
lower_threshold = np.percentile(transfer_fee, 5)

df = detailed_transfers[(detailed_transfers['transferFee'] <= upper_threshold) & (detailed_transfers['transferFee'] >= lower_threshold)]

def normalize_transfer(df):
    for season in df['season'].unique():
        df.loc[df['season'] == season, 'transferFee'] /= season_average[season]

def treat_data(df):
    add_columns = []
    add_columns += dummie_age(df)
    #add_columns += dummie_nationality(df)
    normalize_transfer(df)

    return add_columns




    
```

```python
def normalize_columns(df):
    normalized_df = df.copy()
    for column in df.columns:
        min_val = df[column].min()
        max_val = df[column].max()
        normalized_df[column] = (df[column] - min_val) / (max_val - min_val)
    return normalized_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb

def run_rf(df, n = 100):
    mae = 0
    r2 = 0
    for i in range(n):
        Y = df['transferFee']
        X = df.drop('transferFee',axis=1).fillna(0)

        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

        # Create and fit the linear regression model
        model = RandomForestRegressor()
        model.fit(X_train, y_train)

        # Predict the target values for the test set
        y_pred = model.predict(X_test)

        mae += np.mean(abs(y_pred - y_test))/n

        r2 += r2_score(y_test,y_pred)/n
    print("Mean Absolute Error (MAE):", mae)
    print("R2:", r2)
    return model
```
