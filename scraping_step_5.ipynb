```python
import pandas as pd
```

```python
leagues = ["pl", "ll", "bl", "sa", "l1"]
stats_types = ["standard", "shooting", "passing", "passing_types", "goal_shot_creation", "defense", "possession", "playing_time", "misc"]
```

```python
for type in stats_types:
    dataframes = []
    for league in leagues:
        df = pd.read_csv(f'data/transfers_stats/new/{league}/{type}.csv')
        dataframes.append(df)
    pd.concat(dataframes, axis = 0).to_csv(f'data/transfers_stats/new/all_{type}.csv', index=False)

```
